```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_data1 = pd.read_csv('sample_data/sales_data.csv')
print(test_data1.head())
test_data1.info()
```

```python
revenue_per_product_cat = test_data1.groupby('Product Category')['Total Revenue'].sum().reset_index()
revenue_per_product_cat
```

```python
plt.plot(revenue_per_product_cat['Product Category'], revenue_per_product_cat['Total Revenue'])
```

```python
plt.bar(revenue_per_product_cat['Product Category'], revenue_per_product_cat['Total Revenue'], color="teal")
plt.xticks(rotation=30)
plt.xlabel("Categories")
plt.ylabel("Total Revenue")
plt.title("Total revenue per Product Category")
```

```python
plt.hist(revenue_per_product_cat['Total Revenue'], color='pink' , bins=10)
```

```python
plt.scatter(revenue_per_product_cat['Product Category'], revenue_per_product_cat['Total Revenue'], color="teal", marker="*")
plt.xticks(rotation=30)
```

# Starting Seaborn

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(x="tip", y = "total_bill", data=tips)
plt.title("Scatter plot of total bill over tips")
plt.show()
```

```python
sns.lineplot(x="size", y="total_bill", data=tips)
plt.title("Line plot of size over total bill")
plt.show()
```

```python
sns.barplot(x="day", y="tip", data=tips)
plt.title("Bar plot of tips per day")
plt.show()
```

```python
sns.boxplot(x="day", y="tip", data=tips)
plt.title("Boxplot plot of tips per day")
plt.show()
```

```python
sns.violinplot(x="sex", y="tip", data=tips)
plt.title("Violin plot of tips per Gender")
plt.show()
```

```python
sns.histplot(tips["total_bill"], bins=8, kde=True)
plt.title("Histogram plot of total bill")
plt.show()
```

```python
sns.kdeplot(tips["total_bill"], fill=True)
plt.title("KDE plot of total bill")
plt.show()
```

```python
sns.pairplot(tips)
```

```python
corr = tips[["total_bill","tip","size"]].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap="grey")
```
