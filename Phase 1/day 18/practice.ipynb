```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_data1 = pd.read_csv('sample_data/sales_data.csv')
print(test_data1.head())
test_data1.info()
```

```python
revenue_per_product_cat = test_data1.groupby('Product Category')['Total Revenue'].sum().reset_index()
revenue_per_product_cat
```

```python
plt.plot(revenue_per_product_cat['Product Category'], revenue_per_product_cat['Total Revenue'])
```

```python
plt.bar(revenue_per_product_cat['Product Category'], revenue_per_product_cat['Total Revenue'], color="teal")
plt.xticks(rotation=30)
plt.xlabel("Categories")
plt.ylabel("Total Revenue")
plt.title("Total revenue per Product Category")
```

```python
plt.hist(revenue_per_product_cat['Total Revenue'], color='pink' , bins=10)
```

```python
plt.scatter(revenue_per_product_cat['Product Category'], revenue_per_product_cat['Total Revenue'], color="teal", marker="*")
plt.xticks(rotation=30)
```
